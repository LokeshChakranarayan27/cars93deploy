```python
import pandas as pd
```

```python
df=pd.read_csv('Cars93.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
X=df[['EngineSize','Horsepower','MPG.city','MPG.highway']]
Y=df[['Price']]
```

```python
X.info()
```

```python
xcat=[]
xcon=[]
for i in X.columns:
    if X[i].dtypes==object:
        xcat.append(i)
    else:
        xcon.append(i)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().set_output(transform= 'pandas')
X_pre = scaler.fit_transform(X)
```

```python
import pickle
with open('scaler.pkl','wb') as file1:  # wb is for write binary
    pickle.dump(scaler,file1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X_pre,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
tr_pred=model.predict(xtrain)
ts_pred=model.predict(xtest)
```

```python
with open('model.pkl', 'wb') as file2:
    pickle.dump(model,file2)
```
